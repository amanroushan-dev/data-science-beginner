```python
import pandas as pd
import matplotlib.pyplot as plt

# Small COVID dataset (daily cases for some countries)
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/new_cases.csv"
df = pd.read_csv(url, parse_dates=["date"])

# Pick only India & USA
countries = ["India", "United States"]
subset = df[["date"] + countries]

# Plot
plt.figure(figsize=(10,5))
for country in countries:
    plt.plot(subset["date"], subset[country], label=country)

plt.legend()
plt.title("Daily New COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("Cases")

#  Save the plot
plt.savefig("covid_cases.png", dpi=300)   # saves in repo folder

#  Show the plot
plt.show()
```
